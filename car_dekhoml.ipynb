```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
df =  pd.read_csv("encode.csv")
x=df.drop('price',axis=1)
y=df['price']
x
```

```python
x
```

```python
#  train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

RANDOM FOREST REGRESSOR


```python
rfr = RandomForestRegressor()
model1=rfr.fit(x_train,y_train)
```

```python

rfr_train = model1.predict(x_train)
rfr_test = model1.predict(x_test)
```

```python
import numpy as np
print("train_data_metrics:")
print("R2 score - ",r2_score(y_train,rfr_train))
print("mean absolute error - ",mean_absolute_error(y_train,rfr_train))
print("mean squared error - ",mean_squared_error(y_train,rfr_train))
print("root mean squared error - ",np.sqrt(mean_squared_error(y_train,rfr_train)))
```

```python
print("test_data_metrics:")
print("R2 score - ",r2_score(y_test,rfr_test))
print("mean absolute error - ",mean_absolute_error(y_test,rfr_test))
print("mean_squared_error - ",mean_squared_error(y_test,rfr_test))
print("root mean squared error - ",np.sqrt(mean_squared_error(y_test,rfr_test)))
```

```python
print("test_data_metrics of linear regression:")
print("r2 score - ",r2_score(y_test,lr_test))
print("mean absolute error - ",mean_absolute_error(y_test,lr_test))
print("mean squared error - ",mean_squared_error(y_test,lr_test))
print("root mean squared error - ",np.sqrt(mean_squared_error(y_test,lr_test)))
```

**INSIGHTS**
* From the results, we clearly seen that the model Gradient Boosting algorithm has a very good r2 score of **0.90**.
* The error metrics, what we have is due to
* My target variable is likely in rupees, and the values are very large (in lakhs or crores).
This means:

* Even a 5–10% prediction error on high-priced vehicles can result in errors of ₹1–5 lakhs.

* For example, if a car is worth ₹20 lakhs, even a 5% error = ₹1 lakh → which reflects in your MAE and RMSE.

* Real-World Scenerio:
Predicting housing or car prices often leads to large absolute error values due to the large scale of the prices — and this is not necessarily bad as long as my model's R² is good and the % error is reasonable.



```python
#Hyper parameter tuning  for gradient boosting regressor
n_estimators = [100,150]
learning_rate = [0.01,0.05]
max_depth = [5,7]
min_samples_split = [3,7]
min_sample_leaf = [2,5]
```

```python
parameters = {
    'n_estimators' : n_estimators,
    'learning_rate' : learning_rate,
    'max_depth'    : max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf' : min_sample_leaf
    
}
```

GRIDSEARCHCV

```python
from sklearn.model_selection import GridSearchCV
gscv = GridSearchCV(estimator=gbr,param_grid=parameters,scoring='neg_mean_squared_error',n_jobs=-1,cv=5,verbose=2,error_score='raise')
model =gscv.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
gscv_m=model.best_estimator_
```

```python
gscv_m
```

```python
predictions= gscv_m.predict(x_test)
```

```python
print("test_data_metrics of GridSearch_cv:")
print("r2 score - ",r2_score(y_test,predictions))
print("mean absolute error - ",mean_absolute_error(y_test,predictions))
print("mean squared error - ",mean_squared_error(y_test,predictions))
print("root mean squared error - ",np.sqrt(mean_squared_error(y_test,predictions)))
```

**The model gets a good score of 0.93 after done hyperparameter tuning through gridsearchcv , the model gets 0.90 before doing hyperparameter tuning.**

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    'n_estimators': [50, 100, 150, 250,350],
    'learning_rate': [0.01, 0.1, 0.05, 0.2],
    'max_depth': [5, 7, 9, 15],
    'min_samples_split': [3, 7, 10, 13],
    'min_samples_leaf': [2, 5, 7, 10]
}

rscv = RandomizedSearchCV(
    estimator=gbr,
    param_distributions=param_dist,
    n_iter=25,                      # Try only 25 random combinations
    scoring='neg_mean_squared_error',
    n_jobs=-1,                      # Use all CPU cores
    cv=5,
    verbose=2,
    random_state=42
)

model1 = rscv.fit(x_train, y_train)
```

```python

rscv_m=model1.best_estimator_
predictions1= rscv_m.predict(x_test)
print("test_data_metrics of randomized_search_cv:")
print("r2 score - ",r2_score(y_test,predictions1))
print("mean absolute error - ",mean_absolute_error(y_test,predictions1))
print("mean squared error - ",mean_squared_error(y_test,predictions1))
print("root mean squared error - ",np.sqrt(mean_squared_error(y_test,predictions1)))
```

**The model gets a good score of 0.93, so we took this as a best model to predict the price**
